# Test data generator

```python
# pip install faker
from faker import Faker

num_people = 1000
path = 'testData1000people.json'

nameGenerator = Faker()
names = [nameGenerator.name() for _ in range(num_people)]
print(names)
```

```python
import random

people = []
for i in range(1, num_people + 1):
    cost = [random.randint(0, 100)] + [people[j]['cost'][i] if j < i - 1 else 0 if j == i - 1 else random.randint(0, 100) for j in range(num_people)]
    person = {
        'name': names[i - 1],
        'id': i,
        'cost': cost
    }
    people.append(person)

for person in people[:10]:
    print(person)
```

```python
import json

with open(path, 'w') as file:
    json.dump(people, file)
```

```python
path_dat = "testData1000People.dat"
with open(path_dat, 'w') as file:
    file.write(str(num_people + 1) + "\n")

    file.write("0 ")
    for person in people:
        file.write(str(person['cost'][0]) + " ")
    file.write("\n")

    for person in people:
        for i in person['cost']:
            file.write(str(i) + " ")
        file.write("\n")
```
